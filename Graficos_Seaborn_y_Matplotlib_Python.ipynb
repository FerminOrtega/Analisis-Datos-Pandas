## Importación de librerías
Para este caso, lo que vamos a importar son varias librerías, entre ellas se encuentra Pandas y Numpy (tratamiento de los datos) y para la creación de gráficos vamos a importar la librería de seaborn `import seaborn as sns` y la de matplotlib `import matplotlib.pyplot as plt`

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cargamos los datos con Pandas

```
fichero = "train.csv"
datos = pd.read_csv(fichero, sep=",")
```

```
datos
```

### Pivot table y Plot
A continuación, vamos a mostrar como hacer nustro primer gráfico, para ello vamos a usar la funcion `pivot_table()` en la que le daremos el **index** (eje x) y el **values** (eje y) aplicandole una funcion suma `aggfunc`.

[Info sobre pivot_table()](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html)

La función plot es simplemente la creación de un gráfico sencillo donde podemos estableder el `kind` (tipo de gráfico que queremos).

[Info sobre plot()](https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html)

```
pivote = datos.pivot_table(index='SaleCondition', values='SalePrice', aggfunc=np.sum)
pivote.plot(kind='bar')
```

### Gráfico con estimación de densidad - displot
Esta función es una suma de dos funciones `kdeplot` para mostrar una distribución univariada o bivariada con una estimación de densidad de los datos y `rugplot` que se encarga del gráfico de lineas.

[Info sobre displot()](https://seaborn.pydata.org/generated/seaborn.distplot.html)

Si nos fijamos en el type que aparece encima del gráfico nos damos cuenta de que aparece `matplotlib.axes...` esto es por el parametro que le hemos añadido a la función, mas concretamente `bins` que sirve para limitar los bordes del gráfico.

```
sns.distplot(datos['SalePrice'], bins=100)
```

Aquí tenemos solo la función `kdeplot` por separado

```
sns.kdeplot(datos['SalePrice'], shade=True)
```

### Varios valores en el mismo gráfico y escala de la densidad
Como podemos comprobar para motrar varios datos en un mismo gráfico solo lo que tenemos que hacer es poner las lineas que queremos que se mustren unas debajo de otras

Para hacer que la escala de densisdad varíe tenemos que añadir la opción `bw` en la función que queramos

```
sns.kdeplot(datos['SalePrice'], shade=True, label="Precio")
sns.kdeplot(datos['SalePrice'], shade=True, bw=100, label="Precio bw=100")
sns.kdeplot(datos['SalePrice'], shade=True, bw=1000, label="Precio bw=1000")
```

```
sns.boxplot(x=datos['MSZoning'], y=datos['SalePrice'])
```

```
sns.swarmplot(x=datos['MSZoning'], y=datos['SalePrice'])
```

```
sns.violinplot(x=datos['MSZoning'], y=datos['SalePrice'])
```
